```python
import pandas as pd
import openai
import time
from openai import OpenAI

client = OpenAI(api_key="YOUR_API_KEY") 

# Load your data
df = pd.read_csv("CVS_FILE_PATH_HERE", sep=";")

print(df.head())
```

```python
# Function to clean definitions
def rewrite_definition(name, definition, model="gpt-3.5-turbo"):
    prompt = f"""
You will receive a gesture name and its original definition. Rephrase the definition into a short, clear narrative sentence that concisely describes the gesture, avoiding lists or step-by-step formatting.

Gesture Name: "{name}"
Original Definition: "{definition}"

Rewritten Narrative:
"""
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": prompt}
            ],
            temperature=0.4,
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error on gesture '{name}': {e}")
        return None
```

```python
# Apply rewriting function with basic progress tracking
cleaned_definitions = []
for i, row in df.iterrows():
    print(f"Processing {i+1}/{len(df)}: {row['gesture_name']}")
    cleaned = rewrite_definition(row['gesture_name'], row['gesture_definition'])
    cleaned_definitions.append(cleaned)
    time.sleep(1.2) 
```

```python

# Add the cleaned definitions to dataframe
df['cleaned_definition'] = cleaned_definitions

print(df.head())
```

```python

# Save to Excel
df.to_excel("new_definitions.xlsx", index=False, engine="openpyxl")
```
